On constate que le fichier a bien été ouvert, décortiqué lettre par lettre, et que le tout a été stocké dans une **liste de liste**.

Passsons maintenant au principe d'affichage. Nous allons créer de nouveau une liste, stockée dans la variable `niveau_rect`, qui contiendra la liste des objets rectangles de chaque mur rencontré lors du parcours de la variable `niveau`.
À la création du rectangle, on règlera directement la position du point supérieur gauche à la bonne position, grâce à la fonction `enumerate` qui nous permettra d'avoir le numéro de ligne et le numéro de colonne.

Restera à *blitter* chaque mur à la bonne position à chaque tour de la boucle principale.

```python
try :
    import pygame
    from pygame.locals import * 
    print("Module chargé !")
except :
    print("Module non trouvé !")

    
def Charge_Niveau(fichier):
    with open(fichier,'r') as file :
        return [[c for c in line if c!='\n'] for line in file]

pygame.init() 
#Création de deux variables contenant les dimensions, à savoir 10*72 et 8*72:
Largeur, Hauteur = 10*72, 8*72
fenetre = pygame.display.set_mode((Largeur, Hauteur))

fond=pygame.image.load("fond_ecran_TP2.jpg").convert()
fenetre.blit(fond,(0,0))

mur=pygame.image.load("Mur_SW.png").convert_alpha() # Chargement de l'image d'un mur

niveau=Charge_Niveau('TP2_niv1')##Ici on charge le fichier et on stocke dans la variable niveau.
#print(niveau) Cette ligne est désormais inutile

niveau_rect=[] #On crée une liste vide
for num_line, line in enumerate(niveau): # on recupère le numéro de ligne et une ligne de la variable niveau
    for num_col, caractere in enumerate(line) : #on récupère le numéro de colonne ainsi que le caractère 
        if caractere =='m' : #Si le caractere est 'm' 
            temp=mur.get_rect()# Creation d'un rectangle temporaire
            temp.x, temp.y = num_col*72, num_line*72#règlage des coordonnées
            niveau_rect.append(temp)#le rectangle temporaire est ajouté à la liste
                
                

continuer =1
while continuer ==1 : #tant que continuer est égal à 1, on recommence la boucle
    for evenement in pygame.event.get() : #Pour chaque evenement
        if evenement.type == QUIT :#Si c'est QUIT
            continuer=0#Alors on change la valeur de continuer, et la boucle s'arrête.
            
    fenetre.blit(fond,(0,0))#palcement du fond
    [fenetre.blit(mur,rect) for rect in niveau_rect]#placement des murs
    
    pygame.display.update() #rafraichissement de l'écran


pygame.quit()
```

Là par contre, nous avons un problème ! Il n'y a aucune détection de collisions avec les murs!

Heureusement, Pygame possède des méthodes très pratiques pour détecter ces collisions entre objets `Rect`.
De manière générale, je vous conseille un petit tour sur la page dédiée aux objets `Rect` du manuel de Pygame :
[https://www.pygame.org/docs/ref/rect.html](https://www.pygame.org/docs/ref/rect.html).

Nous utiliserons donc la méthode `collidelist`,qui teste si un objet `Rect` est en collision avec une **liste d'objets `Rect`** passée en argument. Cette fonction renvoie $-1$ si il n'y a pas collision, sinon elle renvoie l'**indice** du rectangle avec lequel il y a collision.

Nous utiliserons donc un objet rectangle temporaire, copie de l'objet `personnage_rect`, modifié en fonction du déplacement souhaité, puis nous testerons la collision avec la liste des rectangles des murs, à savoir `niveau_rect`. Si la valeur renvoyée est égale à $-1$, alors nous modifierons la valeur réelle de `personnage_rect`, sinon nous ne ferons rien.

```python
try :
    import pygame
    from pygame.locals import * 
    print("Module chargé !")
except :
    print("Module non trouvé !")

    
def Charge_Niveau(fichier):
    with open(fichier,'r') as file :
        return [[c for c in line if c!='\n'] for line in file]

pygame.init() 
#Création de deux variables contenant les dimensions, à savoir 10*72 et 8*72:
Largeur, Hauteur = 10*72, 8*72
fenetre = pygame.display.set_mode((Largeur, Hauteur))

fond=pygame.image.load("fond_ecran_TP2.jpg").convert()
fenetre.blit(fond,(0,0))

mur=pygame.image.load("Mur_SW.png").convert_alpha() # Chargement de l'image d'un mur
personnage = pygame.image.load("Faucon/Faucon400.png").convert_alpha()
personnage_rect=personnage.get_rect()
niveau=Charge_Niveau('TP2_niv1')##Ici on charge le fichier et on stocke dans la variable niveau.
#print(niveau) Cette ligne est désormais inutile

niveau_rect=[] #On crée une liste vide
for num_line, line in enumerate(niveau): # on recupère le numéro de ligne et une ligne de la variable niveau
    for num_col, caractere in enumerate(line) : #on récupère le numéro de colonne ainsi que le caractère 
        if caractere =='m' : #Si le caractere est 'm' 
            temp=mur.get_rect()# Creation d'un rectangle temporaire
            temp.x, temp.y = num_col*72, num_line*72#règlage des coordonnées
            niveau_rect.append(temp)#le rectangle temporaire est ajouté à la liste
        if caractere=='d':
            #placement au centre du carré
            personnage_rect.centerx, personnage_rect.centery = num_col*72+36, num_line*72+36
                
                

continuer =1
while continuer ==1 : #tant que continuer est égal à 1, on recommence la boucle
    for evenement in pygame.event.get() : #Pour chaque evenement
        if evenement.type == QUIT :#Si c'est QUIT
            continuer=0#Alors on change la valeur de continuer, et la boucle s'arrête.
    dic_touche=pygame.key.get_pressed()
    rect_temporaire=personnage_rect##Necessaire, sinon rect_temporaire n'existe peut-etre pas...
    if dic_touche[K_DOWN] :
        if personnage_rect.bottom<Hauteur :
            rect_temporaire=personnage_rect.move(0,3)#On met à jour la nouvelle position dans une variable temporaire
            
    if rect_temporaire.collidelist(niveau_rect)==-1 :#si il n'y a pas collision, 
        personnage_rect=rect_temporaire #On met à jour la position du personnage
    fenetre.blit(fond,(0,0)) # placement du fond
    [fenetre.blit(mur,rect) for rect in niveau_rect] #placement des murs
    fenetre.blit(personnage,personnage_rect)
    
    pygame.display.update() #rafraichissement de l'écran


pygame.quit()
```

Et voilà ! En quelques lignes le personnage est bloqué, et les collisions sont bien détectées !
